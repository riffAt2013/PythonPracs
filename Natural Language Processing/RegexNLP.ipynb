## It is very important to learn about regex since we will not always know the exact string value when searching, rather than pattern of the string. regex helps pattern matching and giving relevant informatiion. ver helpful for NLP



```
import re
```

```
search = "Hey monica! my number has changed from 123-456 to 456-789. pls take note of that"
```

```
# the pattern of the phone num is some 3 digits and then an exact hyphen followed by another 3 digits
# since a single digit is repped by \d in regex and a muliple of them takes a {number of digit}
# the pattern then looks as followed 
pattern = r'\d{3}-\d{3}'
re.search(pattern, string = search)
```

```
# store the match object for further use
match = re.search(pattern , string = search)
```

```
#returns a singular instance of the pattern match, but there was two of 'em
match.group()
```

```
re.findall(pattern, search)
```

```
# we might want to take just the iterables for further use not the exact matchings
# finditer() returns match objects just like search() but it creates an iterator
# for multiple matches. very cool!
for i in re.finditer(pattern, search):
  print(i)
```

```
# we can separate individual groups from one single pattern match.
# this will treat first 3 digits of the numbers as first group, 2nd 3 as 2nd group
pattern2 = r'(\d{3})-(\d{3})'
match2 = re.search(pattern2, search)
print(match2.group(1))
print(match2.group(2))
match2.group()
```

```
# the pip operator is another useful tool. it can be used as an OR statement, one pattern for two use case!
pattern3 = r'monica|chandler'
print(re.search(pattern3, search))
re.search(pattern3, string="Chandler is shit!".lower())
```

```
# wildcard character. see this shit!!
re.findall(pattern=r'.at', string = "The cat with a hat sat on a bat flying")
```

```
# '^' stands for starts with
# & for ends with
re.findall(r'\d$', 'Numero uno1')
```

```
# ^ with square brackets tells exclusion
values = re.findall(r'[^\d]+', 'There 3 are asll 2 in the 1 apostle');
''.join(values)
```
